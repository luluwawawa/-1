```python
#1.读取数据
data=pd.read_csv("C:/Users/62678/Desktop/Causal/data/online_classroom.csv")
print(data)
```

```python
#2. 查看数据基本特征(pd.DataFrame)
print(data.shape)
print(data.head(10))
print(data.columns)
print(data.info())
print(data.values)
print(data.describe()) # 查看数据统计特征
```

```python
#3.查看列特征
print(data.dtypes)  #查看每列的数据类型
data['gender'].astype(int)   #修改数据类型

print(data['gender']) # 取列
#loc只能用名索引
print(data.loc[0:2,'gender'])
#iloc只能用数字索引
print(data.iloc[0:2,0:2])
print(data.iloc[[1,2],[1,3]])

# where的用法
print(data.iloc[data['gender']==1])
print(data.iloc[data['gender']!=1&data['asian']==0])

# group_by
data.groupby('gender')
```

```python
A=np.select(
     [data["format_ol"].astype(bool), data["format_blended"].astype(bool)],
     ["online", "blended"],
     default="face_to_face")  # np.select(bool1,bool2,若bool1为真返回值，若bool2为真，其余) --返回323个字符，按照两列来分类
print(A)
B= data.assign(class_format = np.select(
     [data["format_ol"].astype(bool), data["format_blended"].astype(bool)],
     ["online", "blended"],
     default="face_to_face")) # 新增一列class_format
print(B)
C=B.groupby('class_format').mean()
print(C)
```
